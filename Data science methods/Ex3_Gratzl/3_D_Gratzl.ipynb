```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, LpMinimize, lpSum, value


df = pd.read_excel('Exercise 3_data.xlsx')
df = df.dropna(subset = ['Unnamed: 0']) # all rows with NaN in the column 'Unnamed: 0' are droped.

# normalize every column (values between 0 and 1)
c_norm = df.columns[1:]   # exclude the first column, as it is not numerical
df[c_norm] = (df[c_norm] - df[c_norm].min()) / (df[c_norm].max() - df[c_norm].min())

data = df[c_norm].reset_index().drop('index',axis=1) # only data, index starts at zero again.

data

slack_table
fuels = df.iloc[:,0] # names of the fules
X = data.iloc[:, 1:13].values   # this is my data
n_DMUs = X.shape[0]  # number of rows (units)
n_input = X.shape[1]  # number of properties / input variables

scores = []
lambdas = []
slacks = []


# for each unit minimalize the efficiency score z
for k in range(n_DMUs):
    prob = LpProblem(f"DEA_Dual_DMUk_{k+1}", LpMinimize)
    lamda = LpVariable.dicts("lamda", range(n_DMUs), lowBound=0, cat="Continuous")
    slack = LpVariable.dicts("slack", range(n_input), lowBound=0, cat="Continuous")
    z = LpVariable("z", lowBound=None, cat="Continuous")
    prob += z # minimalize z
    
    for i in range(n_input): # constrains: looping over all input variable
        lhs = lpSum([lamda[j] * X[j][i] for j in range(n_DMUs)]) + slack[i]
        rhs = z * X[k][i]
        prob += lhs == rhs, f"InputConstraint_{i+1}"

    prob += lpSum([lamda[j] for j in range(n_DMUs)]) == 1, "Convexity"

    
    prob.solve() # solve the LP problem and store
    eff = value(z)
    scores.append(eff)
    lambda_values = [value(lamda[j]) for j in range(n_DMUs)]
    lambdas.append(lambda_values)
    slacks_values = [value(slack[i]) for i in range(n_input)]
    slacks.append(slack_values)

results_df = pd.DataFrame({
    "DMU": fuels,
    "Efficiency": scores,
    "Efficient": [round(scores[i], 5) == 1.0 and all(round(s, 5) == 0 for s in slacks[i]) for i in range(n_DMUs)]
})

slack_df = pd.DataFrame(slacks, columns=[f"Slack_{i+1}" for i in range(n_input)])
results_df = pd.concat([results_df, slack_df], axis=1)
lambda_df = pd.DataFrame(lambda_table, columns=[f"Lambda_{j+1}" for j in range(n_DMUs)])
lambda_df.insert(0, "DMU", fuels)
results_df = pd.concat([results_df,lambda_df],axis=1)


results_df
```
